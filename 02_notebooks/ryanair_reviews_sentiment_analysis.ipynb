Visualizacion  de datos y selección del modelo.

```python
#Gráficos mas comunes

# Histograma
plt.figure(figsize=(8, 6))
plt.hist(dataset['columna'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Variable')
plt.ylabel('Frecuencia')
plt.title('Histograma de la variable')
plt.show()

# Gráfico de barras
plt.figure(figsize=(8, 6))
sns.countplot(x='columna', data=dataset)
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.title('Gráfico de barras de la variable')
plt.show()

# Diagrama de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(dataset['columna1'], dataset['columna2'], color='green', alpha=0.5)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Diagrama de dispersión entre Variable 1 y Variable 2')
plt.show()

# Gráfico de caja
plt.figure(figsize=(8, 6))
sns.boxplot(x='columna', y='columna2', data=dataset)
plt.xlabel('Variable categórica')
plt.ylabel('Variable numérica')
plt.title('Gráfico de caja de la relación entre Variable categórica y Variable numérica')
plt.show()

# Gráfico de línea
plt.figure(figsize=(8, 6))
plt.plot(dataset['fecha'], dataset['valor'], marker='o', linestyle='-', color='b')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Gráfico de línea de la evolución temporal')
plt.show()

# Mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de calor de la correlación entre variables')
plt.show()
```
